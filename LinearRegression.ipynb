```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('winequality-red.csv', sep=';' , usecols=['alcohol','quality'])
display(dataset)
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values

print(X)
print(Y)
```

```python
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Alcohol vs. Quality')
plt.scatter(X,Y,color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.4)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
print(Y_pred)
print(Y_train)
```

```python
accuracy = model.score(X_test,Y_test)
print(accuracy)
```

```python
plt.scatter(X_train,Y_train,color='blue',marker='+')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Alcohol vs Quality (Train set)')
plt.plot(X_train, model.predict(X_train), color='red')
plt.show()

plt.scatter(X_test,Y_test,color='blue',marker='+')
plt.xlabel('area')
plt.ylabel('price')
plt.title('Area vs. Price (Test Set)')
plt.plot(X_test, model.predict(X_test), color='red')
plt.show()
```

```python
test_data = np.array([[9]])
test_pred = model.predict(test_data)
print(test_pred)
```

```python
AlcoholAmount = float(input("Enter the amount of alcohol that you think is in your drink?"))
QualityOfWine = model.predict(np.array([[AlcoholAmount]])) 
print('Estimated Quality of wine with ' + str(AlcoholAmount) + ' amount of alcohol is ' + str(QualityOfWine[0]))
```
