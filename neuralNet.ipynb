```python
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/Daily_Demand_Forecasting_Orders.csv', sep=';', usecols=['Week of the month','Target (Total orders)'])
display(dataset)
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values

print(X)
print(Y)
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units = 3,activation='relu', input_shape=[1])])
model.add(Dense(1))
print(model.summary())
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
run_model = model.fit(X,Y, epochs=800)
```

```python
print(model.predict([4]))
```

```python
print(run_model.history.keys())
```

```python
plt.plot(run_model.history['loss'])
plt.title('model loss')
```
